## Regionprops is a great way to measure properties of labeled/segmented regions. 

Once you have derived labeled discrete regions (**see the `skimage.measure.label()` function below**), you can measure the properties of those regions and output them to a pandas dataframe. 

For multiple images in a directory, I am guessing that you will need to make a separate dataframe for each image in your dir and then append dataframes each time in the loop. 

```python
from skimage import measure, io, img_as_ubyte
import matplotlib.pyplot as plt
from skimage.color import label2rgb, rgb2gray
import numpy as np
from skimage.filters import threshold_otsu
import pandas as pd
from skimage.measure import label, regionprops
img_path = '/Users/matthewmiller/Desktop/sreeni_github/images/particles.jpg'
image = img_as_ubyte(rgb2gray(io.imread(img_path)))
```

```python
#in this particular case, we can simply 
threshold = threshold_otsu(image)
```

```python
label_image = measure.label(image<threshold,connectivity=image.ndim)

image_label_overlay = label2rgb(label_image,image=image)
```

```python
plt.imshow(image_label_overlay)
```

```python
props = measure.regionprops_table(label_image,image,properties=['label','area','equivalent_diameter','mean_intensity','solidity'])
```

```python
df = pd.DataFrame(props)
```
